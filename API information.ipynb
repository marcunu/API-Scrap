```python
dicc = {}
actividades = []
for i in range(len(response)):
    print()
    dicc = {
        "Nombre carrera": response[i]['name'],
        "Distancia (km)" : round(response[i]['distance']/1000, 2),
        "Tiempo (min)" : round(response[i]['moving_time']/60,2),
        "Pulso medio (ppm)" : response[i]['average_heartrate'],
        "Altitud (m)" : response[i]['elev_low'],
        "Elevacion (m)" : response[i]['total_elevation_gain'],
        "Fecha" : response[i]['start_date'][:10],
        "Ubicacin" : response[i]['start_latlng']
        
    }
    actividades.append(dicc)
```

```python
actividades
```

```python
datos = pd.DataFrame(actividades)
datos.head()
```

```python
locator = Nominatim(user_agent="myGeocoder")
```

```python
coord = (40.46, -3.69)
rgeocode = locator.reverse(coord)
```

```python
info = rgeocode.raw
info
```

```python
info ['address']['country']
```

```python
def ciudad(lon_lat):
    
    coord = lon_lat
    rgeocode = locator.reverse(coord)
    info = rgeocode.raw
    city = info ['address']['city']
    return city

ciudad([40.46, -3.69])
```

```python
lt.ciudad([40.46, -3.69])
```

```python
def comunidad(lon_lat):
    
    coord = lon_lat
    rgeocode = locator.reverse(coord)
    info = rgeocode.raw
    comu = info ['address']['state']
    return comu

comunidad([40.46, -3.69])
```

```python
lt.comunidad([40.46, -3.69])
```

```python
def pais(lon_lat):
    
    coord = lon_lat
    rgeocode = locator.reverse(coord)
    info = rgeocode.raw
    pais = info ['address']['country']
    return pais

pais([40.46, -3.69])
```

```python
lt.pais([40.46, -3.69])
```

```python
datos['Ciudad'] = datos['Ubicacin'].apply(ciudad)
datos['Comunidad Autonoma'] = datos['Ubicacin'].apply(comunidad)
datos['Pais'] = datos['Ubicacin'].apply(pais)
datos.tail()
```

```python
datos['Ritmo medio (min/km)'] = round(datos['Tiempo (min)']/datos['Distancia (km)'],2)
```

```python
datos['Velocidad media (km/h)'] = round(60/datos['Ritmo medio (min/km)'],2)
datos.head()
```

```python
datos.columns
```

```python
carreras = datos[['Nombre carrera', 'Distancia (km)', 'Tiempo (min)',
       'Ritmo medio (min/km)', 'Velocidad media (km/h)', 'Pulso medio (ppm)',
       'Altitud (m)', 'Elevacion (m)', 'Fecha', 'Ciudad',
       'Comunidad Autonoma', 'Pais', 'Ubicacin']]
carreras.sample(5)
```
